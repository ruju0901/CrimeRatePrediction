```python
crime_data['Day of Week'] = pd.to_datetime(crime_data['DATE OCC']).dt.day_name()
day_of_week_groups = crime_data.groupby('Day of Week')
crime_types = ['BATTERY - SIMPLE ASSAULT',
       'VANDALISM - MISDEAMEANOR ($399 OR UNDER)',
       'VANDALISM - FELONY ($400 & OVER, ALL CHURCH VANDALISMS)',
       'RAPE, FORCIBLE', 'SHOPLIFTING - PETTY THEFT ($950 & UNDER)',
       'THEFT-GRAND ($950.01 & OVER)EXCPT,GUNS,FOWL,LIVESTK,PROD']  
crime_type_frequencies = {}
for crime_type in crime_types:
    crime_type_frequencies[crime_type] = day_of_week_groups['Crm Cd Desc'].apply(
        lambda x: (x == crime_type).sum())

plt.figure(figsize=(12, 6))
for crime_type, frequencies in crime_type_frequencies.items():
    plt.plot(frequencies.index, frequencies.values, label=crime_type)

plt.title('Crime Frequency by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Frequency')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(True)
plt.show()
```

```python
grouped_data = crime_data.groupby('AREA NAME')
crime_stats = grouped_data['Crm Cd'].agg(['mean', 'median', 'std'])
plt.figure(figsize=(12, 6))
crime_stats['mean'].sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Mean Crime Rates by Region or City')
plt.xlabel('Region or City')
plt.ylabel('Mean Crime Rate')
plt.xticks(rotation=90)
plt.show()
```

```python
import folium
plt.figure(figsize=(12, 6))
crime_stats['mean'].sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Mean Crime Rates by Region or City')
plt.xlabel('Region or City')
plt.ylabel('Mean Crime Rate')
plt.xticks(rotation=90)

# Create a map to visualize crime rates by region or cit
m = folium.Map(location=[34.0522, -118.2437], zoom_start=10)

for area, mean_rate in crime_stats['mean'].items():
    lat = crime_data.loc[crime_data['AREA NAME'] == area, 'LAT'].mean()
    lon = crime_data.loc[crime_data['AREA NAME'] == area, 'LON'].mean()

    # Customize the marker color based on mean crime rate
    if mean_rate > 100:
        color = 'red'
    else:
        color = 'blue'

    # Create a marker on the map with a pop-up showing the area's crime rate
    folium.CircleMarker(
        location=[lat, lon],
        radius=10,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.6,
        popup=f"{area}: Mean Crime Rate {mean_rate:.2f}"
    ).add_to(m)

# Display the map alongside the bar chart
plt.show()
m.save('crime_map.html')  # Save the map to an HTML file
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create age group categories
age_bins = [0, 18, 30, 45, 60, 100]
age_labels = ['0-18', '19-30', '31-45', '46-60', '61+']
crime_data['Age Group'] = pd.cut(crime_data['Vict Age'], bins=age_bins, labels=age_labels)

# Visualization 1: Age Distribution
plt.figure(figsize=(12, 6))
sns.histplot(data=crime_data, x='Vict Age', bins=20, kde=True, color='skyblue')
plt.title('Age Distribution of Crime Victims')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Visualization 2: Age vs. Crime Type (Separate Bar Charts with Rotated X-Labels and Smaller Fonts)
fig, axes = plt.subplots(2, 3, figsize=(15, 8), sharey=True)
plt.subplots_adjust(wspace=0.1, hspace=0.1)

for i, age_group in enumerate(age_labels):
    ax = axes[i // 3, i % 3]
    subset = crime_data[crime_data['Age Group'] == age_group]
    age_vs_crime = subset['Crm Cd Desc'].value_counts()
    age_vs_crime.plot(kind='bar', ax=ax, color='skyblue')
    ax.set_title(f'Age Group: {age_group}')
    ax.set_xlabel('Crime Types')
    ax.set_ylabel('Frequency')
    ax.set_xticklabels(age_vs_crime.index, rotation=90, fontsize=2)  # Adjust font size and label rotation

plt.tight_layout()
plt.show()

# Visualization 3: Time Trends by Age Group
crime_data['Year'] = crime_data['DATE OCC'].dt.year
time_trends = crime_data.groupby(['Year', 'Age Group'],observed=False).size().unstack().fillna(0)
plt.figure(figsize=(12, 6))
time_trends.plot(kind='line', marker='o', colormap='tab20')
plt.title('Crime Trends Over Time by Age Group')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.legend(title='Age Group', loc='upper left')
plt.show()
```

```python
# Create age group categories
age_bins = [0, 18, 30, 45, 60, 100]
age_labels = ['0-18', '19-30', '31-45', '46-60', '61+']
crime_data['Age Group'] = pd.cut(crime_data['Vict Age'], bins=age_bins, labels=age_labels)

# Visualization 2: Crime vs. Most Common Age Group
most_common_age_group_for_crime = crime_data.groupby(['Crm Cd Desc', 'Age Group'],observed=False).size().unstack().fillna(0)
most_common_age_group_for_crime = most_common_age_group_for_crime.idxmax(axis=1)
most_common_age_group_for_crime = most_common_age_group_for_crime.reset_index(name='Most Common Age Group')

plt.figure(figsize=(12, 6))
sns.barplot(data=most_common_age_group_for_crime, x='Crm Cd Desc', y='Most Common Age Group',hue='Most Common Age Group', legend=False, palette='viridis')
plt.title('Most Common Age Group for Each Crime')
plt.xlabel('Crime Type')
plt.ylabel('Most Common Age Group')
plt.xticks(rotation=90, fontsize=4)
plt.show()
```
